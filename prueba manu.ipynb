Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np


tasa = 2
n=100
vals = np.random.exponential(tasa, n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

```python
df.head()
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt



x=np.linspace(0, max(vals), 1000)
grafica=expon.pdf(x, scale=2)
plt.plot(x, grafica, color='magenta')


plt.title('Tiempos de servicio')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()



```

QQ plot usando statsmodels. 

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data = vals, dist=stats.expon, loc=0, scale=2, line="45")
plt.show()
```

```python
import matplotlib.pyplot as plt
# histograma
count, bins, ignored = plt.hist(x=vals, bins=30, density=True, color='orange')

x=np.linspace(0, max(vals), 1000)
grafica=expon.pdf(x, scale=2)
plt.plot(x, grafica, color='magenta')


plt.title('Tiempos de servicio')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```
